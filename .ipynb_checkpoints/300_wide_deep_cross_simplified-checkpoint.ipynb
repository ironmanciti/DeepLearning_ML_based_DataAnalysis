# 300. Structured data learning with Wide, Deep, and Cross networks

이 예제에서는 두 가지 모델링 기술을 사용하여 구조화된 데이터 분류를 수행하는 방법을 보여줍니다.

1. [Wide & Deep](https://ai.googleblog.com/2016/06/wide-deep-learning-better-together-with.html) models
2. [Deep & Cross](https://arxiv.org/abs/1708.05123) models (Deep & Cross Network for Ad Click Predictions)

## The dataset

- 이 예에서는 UCI Machine Learning Repository의 [Covertype](https://archive.ics.uci.edu/ml/datasets/covertype) ( (주된 삼림 종류)) 데이터 세트를 사용합니다.   

- 작업은 지도 제작 변수(cartographic variables)에서 삼림의 유형을 예측하는 것입니다.  

- 데이터 세트에는 12개의 입력 기능(10개의 숫자 기능 및 2개의 범주형 기능)이 있는 506,011개의 인스턴스가 포함됩니다. 각 인스턴스는 7개의 클래스 중 1개로 분류됩니다.

## Define dataset metadata

여기에서 데이터를 읽고 입력 feature로 구문 분석하고 해당 유형과 관련하여 입력 feature를 인코딩하는 데 유용한 데이터 세트의 메타데이터를 정의합니다.

````python
data.columns
````

````python
TARGET_FEATURE_NAME = "Cover_Type"

TARGET_FEATURE_LABELS = ["0", "1", "2", "3", "4", "5", "6"]

NUMERIC_FEATURE_NAMES = ['Elevation', 'Aspect', 'Slope', 'Hillshade_Noon']

CATEGORICAL_FEATURES_WITH_VOCABULARY = {
    "Soil_Type": list(data["Soil_Type"].unique()),
    "Wilderness_Area": list(data["Wilderness_Area"].unique()),
}

CATEGORICAL_FEATURE_NAMES = list(CATEGORICAL_FEATURES_WITH_VOCABULARY.keys())

FEATURE_NAMES = NUMERIC_FEATURE_NAMES + CATEGORICAL_FEATURE_NAMES

COLUMN_DEFAULTS = [
    [0] if feature_name in NUMERIC_FEATURE_NAMES + [TARGET_FEATURE_NAME] else ["NA"]
    for feature_name in data.columns
]

NUM_CLASSES = len(TARGET_FEATURE_LABELS)

print(len(data.columns), data.columns)
print(len(COLUMN_DEFAULTS), COLUMN_DEFAULTS)
````

## Experiment setup

다음으로, 파일을 읽고 파싱하는 입력 함수를 정의한 다음 학습 또는 평가를 위해 feature와 레이블을 [`tf.data.Dataset`](https://www.tensorflow.org/guide/datasets)로 변환합니다.

````python
def get_dataset_from_csv(csv_file_path, batch_size, shuffle=False):

    dataset = tf.data.experimental.make_csv_dataset(
        csv_file_path,
        batch_size=batch_size,
        column_names=data.columns,
        column_defaults=COLUMN_DEFAULTS,
        label_name=TARGET_FEATURE_NAME,
        num_epochs=1,
        header=True,
        shuffle=shuffle,
    )
    return dataset.cache()
````

## Encode features

input feature에 대한 두 가지 표현인 sparse 및 dense 를 만듭니다.
1. **sparse** 표현에서 categorical feature는 `CategoryEncoding` 레이어를 사용하여 원-핫 인코딩으로 인코딩됩니다. 이 표현은 모델이 특정 feature 값을 *기억*하여 특정 예측을 하는 데 유용할 수 있습니다.
2. **dense** 표현에서 범주형 특징은 '임베딩' 레이어를 사용하여 저차원 임베딩으로 인코딩됩니다. 이 표현은 모델이 보이지 않는 기능 조합을 잘 *일반화*하는 데 도움이 됩니다.

* StringLookup 
    - mask_token : 마스킹된 입력을 나타내는 token. output_mode가 "int"이면 토큰이 어휘에 포함되고 인덱스 0에 매핑. 다른 출력 모드에서는 토큰이 어휘에 나타나지 않고 입력의 마스크 토큰 인스턴스가 삭제됨. None으로 설정하면 마스크 용어가 추가되지 않습니다. 기본값은 ""입니다.
    - output_mode : 레이어 출력에 대한 사양.   
        "int": 입력 토큰의 원시 정수 인덱스를 반환.   
        "binary": 해당 인덱스에 매핑된 토큰이 샘플에 한 번 이상 존재하는 모든 요소에 1을 포함하는 vocab_size 또는 max_tokens 크기의 샘플당 단일 int 배열을 출력합니다.
    - num_oov_indices : 사용할 어휘에 없는 토큰의 수. 이 값이 0이면 output_mode가 "int"일 때 OOV 입력이 -1로 매핑. 

````python
from tensorflow.keras.layers.experimental.preprocessing import StringLookup

vocabulary = CATEGORICAL_FEATURES_WITH_VOCABULARY['Wilderness_Area']
vocabulary
````

````python
test_data = tf.constant([
                ['Wilderness_Area1',
                'Wilderness_Area3',
                'Wilderness_Area4',
                'Wilderness_Area2']])
lookup = StringLookup(
                vocabulary=vocabulary,
                mask_token=None,
                num_oov_indices=0,
                output_mode="int",
            )

print(lookup(test_data).numpy())

lookup = StringLookup(
                vocabulary=vocabulary,
                mask_token=None,
                num_oov_indices=0,
                output_mode="binary",
            )

print(lookup(test_data).numpy())
````

````python
def encode_inputs(inputs, use_embedding=False):
    encoded_features = []
    for feature_name in inputs:
        if feature_name in CATEGORICAL_FEATURE_NAMES:
            vocabulary = CATEGORICAL_FEATURES_WITH_VOCABULARY[feature_name]
            # 문자열 값을 정수 인덱스로 변환하는 lookup 생성.
            # 마스크 토큰을 사용하지 않고 어휘(oov) 토큰을 기대하지 않기 때문에 mask_token을 
            # None으로 설정하고 num_oov_indices를 0으로 설정.
            lookup = StringLookup(
                vocabulary=vocabulary,
                mask_token=None,
                num_oov_indices=0,
                output_mode="int" if use_embedding else "binary",
            )
            if use_embedding:
                # 문자열 입력 값을 정수 인덱스로 변환합니다.
                encoded_feature = lookup(inputs[feature_name])
                embedding_dims = int(math.sqrt(len(vocabulary)))
                # 지정된 차원으로 임베딩 레이어를 생성합니다.
                embedding = layers.Embedding(
                    input_dim=len(vocabulary), output_dim=embedding_dims
                )
                # 인덱스 값을 임베딩 표현으로 변환합니다.
                encoded_feature = embedding(encoded_feature)
            else:
                # 문자열 입력 값을 원 핫 인코딩으로 변환합니다.
                encoded_feature = lookup(tf.expand_dims(inputs[feature_name], -1))
        else:
            # numeric feature를 있는 그대로 사용합니다.
            encoded_feature = tf.expand_dims(inputs[feature_name], -1)

        encoded_features.append(encoded_feature)

    all_features = layers.concatenate(encoded_features)
    return all_features
````

## Create model inputs

이제 모델에 대한 입력을 dictionary로 정의합니다. 여기서 키는 feature name이고 value는 해당 feature shape 및 data type이 있는 `keras.layers.Input` 텐서입니다.

````python
def create_model_inputs():
    inputs = {}
    for feature_name in FEATURE_NAMES:
        if feature_name in NUMERIC_FEATURE_NAMES:
            inputs[feature_name] = layers.Input(
                name=feature_name, shape=(), dtype=tf.float32
            )
        else:
            inputs[feature_name] = layers.Input(
                name=feature_name, shape=(), dtype=tf.string
            )
    return inputs
````

### Train the model

````python
learning_rate = 0.001
dropout_rate = 0.1
batch_size = 265
num_epochs = 50

hidden_units = [32, 32]

def run_experiment(model):

    model.compile(
        optimizer=keras.optimizers.Adam(learning_rate=learning_rate),
        loss=keras.losses.SparseCategoricalCrossentropy(),
        metrics=[keras.metrics.SparseCategoricalAccuracy()],
    )

    train_dataset = get_dataset_from_csv(train_data_file, batch_size, shuffle=True)

    test_dataset = get_dataset_from_csv(test_data_file, batch_size)

    print("Start training the model...")
    history = model.fit(train_dataset, epochs=num_epochs)
    print("Model training finished")

    _, accuracy = model.evaluate(test_dataset, verbose=0)

    print(f"Test accuracy: {round(accuracy * 100, 2)}%")
````

## Experiment 1: a baseline model (- no Deep and Wide)

첫 번째 실험에서 범주형 기능이 원-핫 인코딩되는 다층 피드포워드 네트워크를 생성해 보겠습니다.

````python
hidden_units
````

````python
def create_baseline_model():
    inputs = create_model_inputs()
    features = encode_inputs(inputs)

    for units in hidden_units:
        features = layers.Dense(units)(features)
        features = layers.BatchNormalization()(features)
        features = layers.ReLU()(features)
        features = layers.Dropout(dropout_rate)(features)

    outputs = layers.Dense(units=NUM_CLASSES, activation="softmax")(features)
    model = keras.Model(inputs=inputs, outputs=outputs)
    return model

baseline_model = create_baseline_model()
keras.utils.plot_model(baseline_model, show_shapes=False, rankdir="LR") #horizontal plot.
````

````python
run_experiment(baseline_model)
````

## Experiment 2: Wide & Deep model

두 번째 실험에서는 Wide & Deep 모델을 생성합니다. 모델의 wide part는 선형 모델이고 모델의 deep part는 다층 피드포워드 네트워크입니다.

모델의 wide part에서 입력 feature의 sparse 표현을 사용하고 모델의 deep part에 대한 입력 feature의 dense 표현을 사용합니다.

모든 입력 feature는 서로 다른 표현으로 모델의 두 부분에 모두 기여합니다.

````python
inputs = create_model_inputs()
inputs
````

````python
encode_inputs(inputs)
````

````python
encode_inputs(inputs, use_embedding=True)
````

````python
def create_wide_and_deep_model():

    inputs = create_model_inputs()
    wide = encode_inputs(inputs)

    deep = encode_inputs(inputs, use_embedding=True)
    
    for units in hidden_units:
        deep = layers.Dense(units, activation="relu")(deep)

    merged = layers.concatenate([wide, deep])
    outputs = layers.Dense(units=NUM_CLASSES, activation="softmax")(merged)
    model = keras.Model(inputs=inputs, outputs=outputs)
    return model

wide_and_deep_model = create_wide_and_deep_model()
keras.utils.plot_model(wide_and_deep_model, show_shapes=False, rankdir="LR")
````

````python
run_experiment(wide_and_deep_model)
````

The wide and deep model achieves ~63% test accuracy.

## Experiment 3: Deep & Cross model

세 번째 실험에서는 Deep & Cross 모델을 생성합니다. 이 모델의 딥파트는 이전 실험에서 생성한 딥파트와 동일합니다. Cross 부분의 핵심 아이디어는 Feature Cross의 정도가 레이어 깊이에 따라 증가하는 효율적인 방식으로 명시적 feature crossing을 적용하는 것입니다.

````python
def create_deep_and_cross_model():

    inputs = create_model_inputs()
    x0 = encode_inputs(inputs, use_embedding=True)

    cross = x0
    for _ in hidden_units:
        units = cross.shape[-1]
        x = layers.Dense(units)(cross)
        cross = x0 * x + cross
    cross = layers.BatchNormalization()(cross)

    deep = x0
    for units in hidden_units:
        deep = layers.Dense(units)(deep)
        deep = layers.BatchNormalization()(deep)
        deep = layers.ReLU()(deep)
        deep = layers.Dropout(dropout_rate)(deep)

    merged = layers.concatenate([cross, deep])
    outputs = layers.Dense(units=NUM_CLASSES, activation="softmax")(merged)
    model = keras.Model(inputs=inputs, outputs=outputs)
    return model


deep_and_cross_model = create_deep_and_cross_model()
keras.utils.plot_model(deep_and_cross_model, show_shapes=False, rankdir="LR")
````

````python
run_experiment(deep_and_cross_model)
````

The deep and cross model achieves ~72% test accuracy.

Keras 전처리 레이어를 사용하여 원-핫 인코딩 및 기능 임베딩을 비롯한 다양한 인코딩 메커니즘으로 범주형 기능을 쉽게 처리할 수 있었습니다.  
또한 와이드, 딥 및 크로스 네트워크와 같은 다양한 모델 아키텍처는 다양한 데이터 세트 속성과 관련하여 서로 다른 장점을 가지고 있습니다.  
이들을 독립적으로 사용하거나 결합하여 데이터 세트에 대한 최상의 결과를 얻을 수 있습니다.
